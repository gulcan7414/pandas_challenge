```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

```python
#Calculate the total number of schools

school_name_df= school_data_complete['school_name'].unique()
total_school_df= len(school_name_df)
total_school_df

```

```python
#calculate the total number of students

total_student_df=school_data_complete['student_name'].count()
total_student_df
```

```python
#calculate the total budget
total_budget_df=school_data['budget'].sum()
total_budget_df
```

```python
avg_math_score_df=student_data['maths_score'].mean()
avg_math_score_df
```

```python
#calculate the average reading score
avg_reading_score_df=student_data['reading_score'].mean()
avg_reading_score_df
```

```python
#Calculate the percentage of students with a passing math score(50 or greater)

students_maths_score=student_data.loc[student_data['maths_score']>=50]
number_students_passing_math= students_maths_score['Student ID'].count()
percent_passing_math=(number_students_passing_math/total_student_df)*100
percent_passing_math
```

```python
#Calculate the percentage of students with a passing reading score(50 or greater)
students_reading_score=student_data.loc[student_data['reading_score']>=50]
number_students_passing_reading=students_reading_score['Student ID'].count()
percent_passing_reading=(number_students_passing_reading/total_student_df)*100
percent_passing_reading
```

```python
#Calculate the percentage of students who passed maths and reading(% overall passing)

overall_passing=student_data[(student_data['maths_score']>=50) & (student_data['reading_score']>=50)
                                     ]['Student ID'].count()/total_student_df*100
overall_passing
```

```python
#create a dataframe to hold the above results

summary_df= pd.DataFrame({
    'Total Schools': total_school_df,
    'Total Students': f'{total_student_df:,}',
    'Total Budget': f'$ {total_budget_df:,.2f}',
    'Average Maths Score': f'{avg_math_score_df:,.6f}',
    'Average Reading Score':f'{avg_reading_score_df:,.6f}',
    '% Passing Maths': f'{percent_passing_math:,.6f}',
    '% Passing Reading': f'{percent_passing_reading:,.6f}',
     '% Overall Passing': f'{overall_passing:,.6f}'
}, index=[0])
summary_df 
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
school_summary['Per Student Budget'] = school_summary['Per Student Budget'].replace('[\$,]','',regex=True).astype(float)
school_summary['Spending Ranges(Per Student)'] = pd.cut(school_summary['Per Student Budget'], spending_bins, labels = labels, include_lowest=True)
df_summary = school_summary[['Spending Ranges(Per Student)', 'Average Maths Score', 'Average Reading Score', '% Passing Maths', '% Passing Reading', '% Overall Passing']].reset_index(drop=True)
spending_summary = df_summary.groupby('Spending Ranges(Per Student)').mean()
spending_summary
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
school_summary['School Size'] = pd.cut(school_summary['Total Students'], size_bins, labels = labels, include_lowest=True)
df_size_summary = school_summary[['School Size', 'Average Maths Score', 'Average Reading Score', '% Passing Maths', '% Passing Reading', '% Overall Passing']].reset_index(drop=True)
size_summary = df_size_summary.groupby('School Size').mean()
size_summary
```
